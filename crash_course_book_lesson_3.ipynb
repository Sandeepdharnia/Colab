## Sorting the list
sort() -> using sort() method we sort a list permanently and cant revert back to its original order. 
a_list = [2, 6, 4]
a_lsit.sort()
a_list output [2, 4, 6]

if we want our list in original order and time to time we need it as sorted then sorted() method is quite handy.

a_list = [2, 6, 4]
sorted(a_list) output [2, 4, 6]
a_list output [2, 6, 4]

```python
# sort a list permanently with sort method
fruit_list = ['Banana', 'Apple', 'Kiwi', 'Cherry']
print(fruit_list)
fruit_list.sort()
print(fruit_list)
fruit_list.sort(reverse=True)
print(fruit_list)
```

```python
# sort a list temporarly  with sorted method
fruit_list = ['Banana', 'Apple', 'Kiwi', 'Cherry']
print(sorted(fruit_list))
print(fruit_list)
```

## Reverse order of the list
reverse() -> this method will reverse the order of list. The reverse impact will be permanently. But will not sort it. 

a_list = [2, 6, 4]

a_list.reverse()

a_list output [4, 6, 2]




```python
#list as reverse order 
fruit_list = ['Banana', 'Apple', 'Kiwi', 'Cherry']
print(fruit_list)
fruit_list.reverse()
print(fruit_list)
```

```python
# exersice 3-8

#Store the locations in a list. Make sure the list is not in alphabetical order.

countries = ['India', 'Ghana', 'Peru', 'Italy']
print(countries)

#Print your list in its original order. Don’t worry about printing the list neatly, just print it as a raw Python list.

#Use sorted() to print your list in alphabetical order without modifying the actual list.
print(sorted(countries))

#Show that your list is still in its original order by printing it.
print(countries)

#Use sorted() to print your list in reverse alphabetical order without changing the order of the original list.”
print(sorted(countries, reverse=True))
```

```python
#Show that your list is still in its original order by printing it again.
print(countries)

#Use reverse() to change the order of your list. Print the list to show that its order has changed.
countries.reverse()
print(countries)
#Use reverse() to change the order of your list again. Print the list to show it’s back to its original order.
countries.reverse()
print(countries)

#Use sort() to change your list so it’s stored in alphabetical order. Print the list to show that its order has been changed.
countries.sort()
print(countries)
#Use sort() to change your list so it’s stored in reverse alphabetical order. Print the list to show that its order has changed.”
countries.sort(reverse=True)
print(countries)
```
