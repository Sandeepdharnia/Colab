## If statments 
to examin a set of condtions and deciding which action to take based on those conditions 

```python
cars = ['bmw', 'toyota', 'suzuki']

for car in cars:
  if car == 'bmw':  # note the :ternary if: print(car.upper() if car == 'bmw' else car.title())
    print(car.upper())
  else:
    print(car.title())
```

```python
# checking for equality
# car is assigned to Bmw, assigning take single equal sign '=' 

car = 'Bmw'

# Now checking that the car is equal to 'Bmw' output will be True. 
# checking equality takes double equal sign '==' 

car == 'Bmw'
```

```python
car == 'Audi'
```

```python
# Ignoring case when checking for equality
# Testing for equality is case sensitive

car = 'Audi'

car == 'audi'
```

```python
# if case does not matter we can chage the case

car = 'Audi'

car.lower() == 'audi'
```

```python
# checking for Inequality 
# Using excalmation point and equal sign (!=) means not equal 

car = 'audi'

if car != 'bmw':
  print(f"We dont sell {car.title()}")
```

##If condition and if-elif-else

if condition check all possibilities 
but if-elif-else condition block checks
one condition after another,

if one of condition fulfilled then print the line. 

```python
# In this case, the condition works on either this or that condition but 
# only one condition will be fulfilled. 
# Once the condition if satisfied then no further run tests 

requested_topping = ["mushrooms", "extra cheese"]

if "mushrooms" in requested_topping:
  print("Adding mushrooms")
elif "pepperoni" in requested_topping:
  print("Adding pepperoni")
elif "extra cheese" in requested_topping:
  print("Adding extra cheese")
else:
  print("No extra topping")
```

### Checking that a list is not empty.

```python
# simple example of nested if statements
# It will check if list is empty or not then procced to next step
requested_toppings = []

if requested_toppings:
  for requested_topping in requested_toppings:
    if requested_topping == "green peppers":
      print(f"Sorry, we are out of {requested_topping} right now")
    else:
      print(f"Adding {requested_topping}")
else:
  print(f"Are you sure you want a plain pizza?")
print("\nYour pizza is just to finish!")
```
