## Dictionaries
a dictionary is a collection of key-value pairs. Each key is connected to a value and can use a key to access the value associated with that key.

A key's value can be a number, a string, a list, oe even another dictionary. In fact, you can use any object that you can create in python as a value in a dictionary.


```python
# simple dictionary
alien_0 = {"color":"green",
           "point": 5}

print(alien_0["color"])
print(alien_0["point"])
```

```python
alien_0 = {"color":"green", 
           "point":5}

print(alien_0["color"])
print(alien_0["point"])
```

```python
# accessing the point value 
new_point = alien_0["point"]
print(new_point)
print(f"You just earned {new_point} points!")
```

```python
alien_0 = {'x':0, 'y':25, 'speed':'medium'}

print(f"Original position: {alien_0['x']}")

# Move the alien to right.
# Determine how far to move the alien based on it's current speed.
if alien_0['speed'] == 'slow':
  x_increment = 1
elif alien_0['speed'] == 'medium':
  x_increment = 2
else:
  x_increment = 3

alien_0['x'] = alien_0['x'] + x_increment 
print(alien_0['x'])
```

### Get method 
get() requires a key as a first argument. 

As a second optional argument, you can pass the value to be returned if the key doesnt exist.

```python
alien_0 = {'color':'red', 
           'speed':'slow'}
# Here dictionary does not have 'point' key and its value
point_value = alien_0.get('point', 'No point value assigned.')
print(point_value)

alien_0['point'] = 15

# Above point key assigned and has a value.

point_value = alien_0.get('point', 'No point value assigned.')
print(point_value)
```

```python
## Nesting 

# Sometimes you will want to store multiple dictionaries in a list, 
# or a list of items as a value in a dictionary. This is called nesting.

# A list of dictionaries

alien_0 = {'color': 'green', 'point':5}
alien_1 = {'color':'yellow', 'point':10}
alien_2 = {'color':'red', 'point':15}

aliens = [alien_0, alien_1, alien_2]
for alien in aliens:
  print(alien)
```

```python
aliens = []

for alien_number in range(30):
  new_alien = {'color': 'green', 'point':5, 'speed':'slow'}
  aliens.append(new_alien)

# show first 5 aliens

for alien in aliens[:5]:
  print(alien)

# show how many aliens have been created

print(f"Total number of aliens: {len(aliens)}")
```

```python
import collections
# exercise 6-10

favorite_numbers = {
    'Raj': [2,4,7],
    'Nina': [7,3],
    'Piku': [9,2],
    'Roni': [6,9,4],
    'Kala': [99, 12, 10],
}

shared_values = collections.defaultdict(list)
for name, values in favorite_numbers.items():
  #print(f"{name.title()} and favorite number(s) are")
  for value in values:
    if value not in shared_values:
      shared_values[value] = []
    shared_values[value].append(name)


for number, shared_names in shared_values.items():
  print(f"{number} is common between: ")
  print(f"\t{shared_names}")
```
